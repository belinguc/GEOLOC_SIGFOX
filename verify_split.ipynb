## PATH

## Load new csv

```python
# df_mess_train
df_mess_train = pd.read_csv(path_data + 'mess_train_list.csv')
print("Shape df_mess_train : ", df_mess_train.shape)


# train_X
train_X = pd.read_csv(path_data + 'train_X.csv')
print("Shape train_X : ", train_X.shape)

# train_y
train_y = pd.read_csv(path_data + 'train_y.csv')
print("Shape train_y : ", train_y.shape)

# val_X
val_X = pd.read_csv(path_data + 'val_X.csv')
print("Shape val_X : ", val_X.shape)

# val_y
val_y = pd.read_csv(path_data + 'val_y.csv')
print("Shape val_y : ", val_y.shape)

# test_X
test_X = pd.read_csv(path_data + 'test_X.csv')
print("Shape test_X : ", test_X.shape)
```

## Plot density

```python
def plot_density(arg):
    df_mess_train[arg].plot.density(label="train_total")
    train_X[arg].plot.density(label="train_X")
    val_X[arg].plot.density(label="val_X")
    test_X[arg].plot.density(label="test_X")

    plt.title(arg)
    plt.legend()
    plt.show()
```

### Plot RSSI density
> **rssi**: On a les mêmes densité pour les train et test sets.

```python
plot_density('rssi')
```

### Plot Nseq density
> **nseq**: fréquence à laquelle les messages sont envoyés (contient peu d'info)

```python
plot_density('nseq')
```

### Plot Bsid density

```python
plot_density('bsid')
```
